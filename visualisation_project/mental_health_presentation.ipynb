```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
year_percent = pd.read_csv("year_percent.csv")
cgpa_percent = pd.read_csv("cgpa_percent.csv")
gender_percent = pd.read_csv("gender_percent.csv")
df = pd.read_csv("mental_health.csv")
```

# 63.37% of university students said that they suffer from mental health issues

```python
sns.lineplot(x="year_of_study", y="depression", data=year_percent, color="teal")
plt.xlabel("Year of Study")
plt.ylabel("% Students with Depression")
plt.ylim(0, 50)
plt.suptitle("Percentage Amount of Univeristy Students")
plt.title("that Suffer from Depression by Year of Study")
plt.show()
```

```python
sns.lineplot(x="year_of_study", y="anxiety", data=year_percent, color="green")
plt.xlabel("Year of Study")
plt.ylabel("% Students with Anxiety")
plt.ylim(0, 50)
plt.suptitle("Percentage Amount of Univeristy Students")
plt.title("that Suffer from Anxiety by Year of Study")
plt.show()
```

```python
sns.lineplot(x="year_of_study", y="panic_attack", data=year_percent, color="firebrick")
plt.xlabel("Year of Study")
plt.ylabel("% Students that had Panic Attacks")
plt.ylim(0, 50)
plt.suptitle("Percentage Amount of Univeristy Students")
plt.title("that Suffer from Panic Attacks by Year of Study")
plt.show()
```

```python
ax = sns.countplot(df['year_of_study'], color="darkviolet")
plt.xlabel("Year Of Study")
plt.ylabel("Total in Survey")
plt.title("Total amount that took survey based on year of study")
for p in ax.patches:
   ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.33, p.get_height()+0.01))
plt.show()
```

```python
sns.barplot(x="CGPA", y="depression", data=cgpa_percent, color="teal")
plt.xlabel("CGPA")
plt.ylabel("% Students with Depression")
plt.suptitle("Percentage Amount of Univeristy Students")
plt.title("that Suffer from Depression by CGPA")
plt.show()
```

```python
sns.barplot(x="CGPA", y="anxiety", data=cgpa_percent, color="green")
plt.xlabel("CGPA")
plt.ylabel("% Students with Anxiety")
plt.suptitle("Percentage Amount of Univeristy Students")
plt.title("that Suffer from Anxiety by CGPA")
plt.show()
```

```python
sns.barplot(x="CGPA", y="panic_attack", data=cgpa_percent, color="firebrick")
plt.xlabel("CGPA")
plt.ylabel("% Students that had Panic Attacks")
plt.suptitle("Percentage Amount of Univeristy Students")
plt.title("that Suffer from Panic Attacks by CGPA")
plt.show()
```

```python
cgpa_order = ['3.50 - 4.00', '3.00 - 3.49', '2.50 - 2.99', '2.00 - 2.49', '0 - 1.99']
ax = sns.countplot(df['CGPA'], order=reversed(cgpa_order), color="darkviolet")
plt.xlabel("CGPA")
plt.ylabel("Total in Survey")
plt.title("Total amount that took survey based on CGPA")
for p in ax.patches:
   ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.33, p.get_height()+0.01))
plt.show()
```
