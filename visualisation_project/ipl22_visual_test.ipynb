```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
bat = pd.read_csv("batting.csv")
bat.head()
```

```python
bat.describe()
```

```python
bat.max()
```

```python
team = bat.groupby("Team")
```

```python
team.mean()
```

```python
team.head()
```

```python
top_bat = bat[["Runs", "Balls", "PlayerName"]].groupby("PlayerName").sum().sort_values("Runs", ascending=0)
top_bat.head(10)
```

```python
top_bat["Economy"] = top_bat["Runs"] / top_bat["Balls"]
top_bat.head(10)
```

```python
top_bat = top_bat.sort_values("Economy", ascending=0)
top_bat.head(10)
```

```python
sns.barplot(data=top_bat, x="PlayerName", y="Economy")
```
